```python
# Q2 Find simple interest
principle = 1000
rate = 5
time = 2
simple_interest = (principle* rate* time) / 100
print("Simple Interest:" , simple_interest)
```

```python
# Q3 Convert temperature from celsius to fahrenheit
celsius = 30
fahrenheit = (celsius * 9/5) + 32
print("Temperature in fahrenheit:" , fahrenheit)
```

```python
# Q4 calculate average of 3 numbers 
x = 10
y = 20
z = 30
average = (x + y + z) / 3
print("Average:" , average)
```

```python
# Q4 find square and cube of a number
a = 10
square = a ** 2
b = 10
cube = b ** 3
print ("Square of a number:" , square)
print("Cube of a number:" , cube)
```

```python
# Q5 Swap two numbers without third variable 
x = 20
y = 40
x , y = y , x
print("After swapping:")
print("X =" , x)
print("Y =" , y)
```

```python
# Q6 Create a Student Report Program that take student details using input(), 
# Store marks in variables, Calculate total and percentage , Use comments, Use proper indentation
name = input("Enter Student Name:")
age = input("Enter age:")

# Marks of 3 subjects 
marks_1 = int(input("Enter subject 1 marks:"))
marks_2 = int(input("Enter subject 2 marks:"))
marks_3 = int(input("Enter subject 3 marks:"))

# calculate total and percentage
total = marks_1 + marks_2 + marks_3
percentage = (total / 300) * 100

# report
print ("Student Report Program:")
print("Name:" , name)
print("Age:" , age)
print("Total:" , total)
print("percentage:" , percentage , "%")

```
