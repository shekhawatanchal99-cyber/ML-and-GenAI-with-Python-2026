```python
# 1. Find sum of first 10 natural numbers.
total_sum = sum(range(1,11))
print("Sum of first 10 natural numbers:" , total_sum)

# another way
total_sum = 0
for i in range(1,11):
    total_sum += i
    print("Sum of first 10 natural numbers:" , total_sum)
```

```python
# 2. find factorial of a number
num = 5
factorial = 1
for i in range(1, num + 1):
    factorial *= i
    print("factorial :" ,factorial)
```

```python
# 5. Create Student Result System

# input student details
name = input("Enter your Name: ")
roll_no = int(input("Enter roll no:"))

# input marks
marks_1 = int(input("Enter subject 1 marks: ")) 
marks_2 = int(input("Enter subject 2 marks: ")) 
marks_3 = int(input("Enter subject 3 marks: ")) 
marks_4 = int(input("Enter subject 4 marks: "))

# Calculate percentage 
total = marks_1 + marks_2 + marks_3 + marks_4
percentage = (total / 400)*100

# Display grade 
if percentage >= 90:
    grade = "A+"
elif percentage >= 80:
    grade = "A"
elif percentage >= 70:
    grade = "B"
elif percentage >= 60:
    grade = "C"
elif percentage >= 50:
    grade = "D"
else:
    grade = "F"

print( " Student Result:")
print("Name:" ,name)
print("Roll no:" , roll_no)
print("Total marks:" , total)
print("Percentage:", percentage)
print("Grade:" , grade)


```
