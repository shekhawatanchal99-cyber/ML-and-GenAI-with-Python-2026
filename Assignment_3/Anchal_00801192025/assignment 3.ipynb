```python
# print first 10 natural numbers
def print_natural():
    for i in range(1, 11):
        print(i)
print_natural()
```

```python
# sum of first N natural numbers
def sum_natural(n):
    return n * (n + 1) // 2
n = 10
print("Sum =" , sum_natural(n))
```

```python
# reverse a number
def reverse_number(num):
    rev = 0
    while num > 0:
        digit = num % 10
        rev = rev * 10 + digit
        num //= 10
    return rev
num = 241
print("Reverse = " , reverse_number(num))
    
```

```python
# count digits in a number
def count_digit(num):
    count = 0
    while num > 0:
        count += 1
        num //= 10
    return count
num = 111
print("Digits = " , count_digit(num))    
```

```python
# check palindrome number
def palindrome(num):
    original = num
    rev = 0
    while num> 0:
        digit = num % 10
        rev = rev * 10 + digit
        num //= 10
    if original == rev:
        return "palindrome" 
    else:
        return "not palindrome"
num = int(input("Enter number : "))
print(palindrome(num))    
```

```python
# generate fibonacci series
def fibonacci(n):
    a, b = 0,1
    print("fibonacci series:")
    for i in range(n):
        print(a, end=" ")
        a, b = b, a+b
n = 10
fibonacci(n)
```

```python
# Calculator Using Functions that contains the following features;

def add(a, b):
    return a + b
def subtract(a, b):
    return a - b
def multiply(a, b):
    return a * b	
def divide(a, b):
    if b ==0:
        return "cannot divide by zero"
    return a / b
print("1. Addition")
print("2. subtraction")
print("3. multiplication")
print("4. division")

choice = int(input("Enter your choice:"))
a = float(input("Enter first number:"))
b = float(input("Enter second number:"))
if choice == 1:
    print("result = " , add(a, b))
elif choice ==2:
    print("result = " , subtract(a, b))
elif choice ==3:
    print("result = " , multiply(a, b))
elif choice ==4:
    print("result = " , divide(a, b))
else:
    print("invalid choice")
```

```python
# handle division by zero using exception
try:
    num1 = int(input("enter first number: "))
    num2 = int(input("enter second number: "))

    result = num1 / num2
    print("Result =", result)

except ZeroDivisionError:
    print("error: division by zero is not allowed.")

except ValueError:
    print("error: enter valid numbers.")
```

```python
# create a student class with name and marks
class Student:

    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def display(self):
        print("student Name:", self.name)
        print("marks:", self.marks)

# Creating object
name = input("enter Student Name: ")
marks = int(input("enter Marks: "))

s1 = Student(name, marks)
s1.display()
```
